```python
#Exceptions from Mohler https://aclanthology.org/L16-1668/:
exs = ['брать', 'быть', 'взять', 'давать', 'дать', 'делать', 'мочь', 'находиться', 
       'оказаться', 'оказываться', 'передавать', 'передать', 'подорожать', 'получать',
       'получить', 'сделать', 'смочь', 'составлять', 'состоять', 'являться']
#My exceptions added:
exs = exs + ['нет', 'бывать', 'иметь']
```

```python
def procfile(fn):
  df = pd.read_csv(foo+fn)
  fl=0
  info = {'verb':[], 'lemma':[], 'sent':[]}
  verbs, lemmas, sents = [], [], []
  sent = []
  for i, r in df.iterrows():
    sent.append(r.Word)
    if (r.POS == 'VERB') and (not (r.Lemma in exs)):
      verbs.append(r.Word)
      lemmas.append(r.Lemma)
    elif r.Word in ['.', '!', '?', '...']:
      sents= len(verbs) * [' '.join(sent)]
      info['verb']+=verbs
      info['lemma']+=lemmas
      info['sent']+=sents
      #print(info)
      verbs, lemmas, sent = [], [], []
  pd.DataFrame(info).to_csv(resfoo+fn)
  return info
```

```python
fns1 = os.listdir(resfoo)
print(len(fns))
for i, fn in enumerate(fns):
  if i%20 == 0:
    print(i, fn)
  #if fn in fns1:
  #  continue
  procfile(fn)
```
