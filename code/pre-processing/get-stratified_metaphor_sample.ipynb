```python
import os, shutil
import pandas as pd
from sklearn.model_selection import StratifiedKFold
```

```python
foo = './drive/MyDrive/rscf/rusidiolect_data/raw/'
fn = 'fpi_1045+ids.tsv'
df = pd.read_csv(foo+fn, sep = '\t', index_col=0)
df = df[df['Автор: Пол'] != 'не указан']
df
```

```python
df['sexgenre'] = df['Автор: Пол'] + '_' + df['Текст: Тип текста']
df
```

```python
df.sexgenre.value_counts()
```

```python
skf = StratifiedKFold(n_splits=10)
folds = []
for train_index, test_index in skf.split(df.index, df['sexgenre']):
  folds.append(df.index[test_index])
folds[:2]
```

```python
df.loc[folds[0]].sexgenre.value_counts()
```

```python
df.loc[folds[1]].sexgenre.value_counts()
```
