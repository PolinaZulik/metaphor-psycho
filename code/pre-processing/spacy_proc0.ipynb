```python
lemmapping = {'меня':'я', 'мне': 'я', 'мной': 'я',
           'тебя':'ты', 'тебе': 'ты', 'тобой': 'ты',
            'ему': 'он', 'ним': 'он', 'нем': 'он',
           'ней': 'она', 'ей': 'она', 
           'нас': 'мы', 'нам': 'мы', 'нами': 'мы',
           'вас': 'вы', 'вами': 'вы', 'вам': 'вы', 
           'им': 'они', 'ними': 'они', 'них': 'они'}
lemmappingPoss = {'его': 'он', 'ее': 'она', 'её': 'она', 'их': 'они'}
```

```python
def proctext(t):
  doc = nlp(t)
  res = []
  for token in doc:
    cn = [x.idx for x in token.children]
    an = [x.idx for x in token.ancestors]
    lemma = token.lemma_
    if not (token.dep_ in ['det', 'nmod']):
      lemma = lemmappingPoss.get(token.lemma_, token.lemma_)
    lemma = lemmapping.get(token.lemma_, lemma)
    #print(token.idx, token.text, token.lemma_, token.pos_, token.dep_, cn)
    res.append([token.idx, token.text, lemma, token.pos_, token.dep_, an, 
                cn, str(token.morph)])
  return res
```

```python
foo = './drive/MyDrive/rscf/rusidiolect_data/'
def proccorpus(corpusname = 'big5_536'):
  fn = foo + 'raw/' + corpusname + '.tsv'
  fn1 = fn.replace(corpusname, corpusname+'+ids')
  df = pd.read_csv(fn, sep = '\t')
  old_author = 0
  x=0
  tnames = []
  for r in df.iterrows():
    author, autage, autsex, ttype, t = [r[1][df.columns[j]] for j in [7, 4, 8, 6, 5]]
    res = proctext(t)
    ###Ниже формирование ключей-имен файлов из таблицы .tsv типа как в этой папке https://drive.google.com/drive/folders/1fKORz7LjkFos9Yq9HnG6Ld9ZcAhbhImH
    corp = re.sub('[a-zа-я ]+', '', r[1][df.columns[11]])
    fauthor = str(author)+corp
    if old_author == author:
      x += 1
    else:
      x = 0
    
    fname = '_'.join([str(x) for x in [fauthor, autage, autsex[0], x, ttype]])
    ###
    tnames.append(fname)
    fname = fname + '.csv'
    data = pd.DataFrame([v[1:] for v in res], index = [v[0] for v in res], 
              columns = ['Word', 'Lemma', 'POS', 'Dep-type', 'Ancestors', 'Dependants', 'Morph'])
    data.to_csv(foo+'processed_spacy/' + corpusname + '/'+fname)
    old_author = author
    if author % 50 == 0:
      print(author, 'ready')
  df['TextId'] = tnames
  df.set_index('TextId').to_csv(fn1)
```
