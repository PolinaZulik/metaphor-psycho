```python
def aggregate3(res4):
  if sum(res4) > 2:
    return 1
  elif sum(res4) < 2:
    return 0
  #else:
  #  return np.nan
```

```python
res['aggregate3of4'] = res.apply(lambda x: aggregate3([x.prediction, x.Polina_metaphor, x.Victoria_metaphor, x.Elena_metaphor]), axis=1)
res
```

```python
res.to_csv('./drive/MyDrive/metaphor-psycho/metaphor_annotation_fpi/res_LCC+Yulia_trained_fullinfo.csv')
```

```python
textfoo = './drive/MyDrive/rscf/rusidiolect_data/processed_spacy/fpi_1045/'
stat = {}
for tid in res.index.unique():
  text = res.loc[tid].dropna()
  stat[tid] = {'agreed_verbs':text.shape[0], 'metaphors':text.aggregate3of4.sum()}
  #print(tid)
  df = pd.read_csv(textfoo+tid+'.csv')
  stat[tid]['text_length'] = df.shape[0]
  stat[tid]['verbs'] = df[df.POS == 'VERB'].shape[0]
stat = pd.DataFrame(stat).transpose()
stat.to_csv('./drive/MyDrive/metaphor-psycho/stats_ver0.csv')
```

```python
stat['metaphors_ratio_agreedverbs'] = stat.metaphors/stat.agreed_verbs
stat['metaphors_ratio_verbs'] = stat.metaphors/stat.verbs
stat['metaphors_ratio_textlength'] = stat.metaphors/stat.text_length
stat
```

```python
all = stat.join(adf1)
all['sex'] = all['Автор: Пол'].apply(lambda x: 0 if x == 'женский' else 1)
all.shape
```

```python
all = all[['verbs', 'metaphors', 'metaphors_ratio_agreedverbs', 
           'metaphors_ratio_verbs', 'metaphors_ratio_textlength', 'sex', 
       'Фрайбургский многофакторный личностый опросник: Невротичность',
       'Фрайбургский многофакторный личностый опросник: Спонтанная агрессивность',
       'Фрайбургский многофакторный личностый опросник: Депрессивность',
       'Фрайбургский многофакторный личностый опросник: Раздражительность',
       'Фрайбургский многофакторный личностый опросник: Общительность',
       'Фрайбургский многофакторный личностый опросник: Уравновешенность',
       'Фрайбургский многофакторный личностый опросник: Реактивная агрессивность',
       'Фрайбургский многофакторный личностый опросник: Застенчивость',
       'Фрайбургский многофакторный личностый опросник: Открытость',
       'Фрайбургский многофакторный личностый опросник: Экстраверсия - интраверсия',
       'Фрайбургский многофакторный личностый опросник: Эмоциональная стабильность',
       'Фрайбургский многофакторный личностый опросник: Маскулинизм - феминизм']]

all.dropna(inplace=True)
```

```python
resd = pd.DataFrame()
for c in all.columns[:5]:
   r = all[all.columns[5:]].corrwith(all[c], method = stats.pearsonr)
   resd[c + ' Corr'] = r.loc[0]
   resd[c + ' P-Value'] = r.loc[1]
resd.index = resd.index.map(lambda x: x.replace('Фрайбургский многофакторный личностый опросник: ', ''))
resd
```
