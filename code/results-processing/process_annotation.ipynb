```python
#fn = fns[0]
def read_annot(df):
  df.index = df.index.astype(np.int16)
  for col in df.columns:
    if col.endswith('metaphor'):
      df[col].fillna(np.nan, inplace=True)
      #df[col] = df[col].astype('Int64')
      #print(df[col])
  return df
```

```python
Lena = []
Vika = []
Polina=[]
sum = 0
lack = []
uneq = 0
verbs = []
for fn in fns:
  df = pd.read_excel(foo1+fn, index_col=0)
  #print(fn)
  if ('Elena_metaphor' in df.columns) and ('Victoria_metaphor' in df.columns)  and ('Polina_metaphor' in df.columns):
    df = read_annot(df)
    for x in df.sent:
      if 'ругай!' in x:
        print(fn)
    sum+=1
    if not df.Elena_metaphor.equals(df.Victoria_metaphor):
      #print(fn)
      uneq += 1
      verbs += list(df[df.Victoria_metaphor != df.Elena_metaphor].lemma)
    Lena += list(df.Elena_metaphor)
    Vika += list(df.Victoria_metaphor)
    if df.Elena_metaphor.hasnans or df.Victoria_metaphor.hasnans or df.Polina_metaphor.hasnans :
      print(fn)
    if ('Polina_metaphor' in df.columns):
      Polina += list(df.Polina_metaphor)
    #else:
    #  print(fn)
  else:
    lack.append(fn)
sum, uneq, len(Lena), len(Vika), len(Polina)
```
